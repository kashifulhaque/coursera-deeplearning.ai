### **Function with one global minimum**

- $f(x) = e^{x} - log(x)$ defined for $x \gt 0$
    - Function in one variable which has only one **minimum point (global minimum)**
    - Sometimes, the minimum point can't be found analytically by solving the equation $\frac{df}{dx} = 0$
    - It can be done using Gradient Descent

- To implement gradient descent
    - Start at some initial point $x_{0}$
    - Aim to find a point where derivative equals to $0$
    - Move down the gradient slope

$$x_1 = x_0 - \alpha \frac{df}{dx}(x_0)\tag{1}$$

- $\alpha \gt 0$ is a paramter called **learning rate**
- Repeat the process iteratively as the number of iterations $n$ is also a parameter

First, define function $f\left(x\right)=e^x - \log(x)$ and its derivative $\frac{df}{dx}\left(x\right)=e^x - \frac{1}{x}$:

```python
def f_example_1(x):
    return np.exp(x) - np.log(x)

def dfdx_example_1(x):
    return np.exp(x) - 1 / x
```

```python
plot_f([0.001, 2.5], [-0.3, 13], f_example_1, 0.0)
```

```python
def gradient_descent(dfdx, x, learning_rate = 0.1, num_iterations = 100):
    for i in range(num_iterations):
        x = x - learning_rate * dfdx(x)
    
    return x
```

```python
num_iterations = 25
learning_rate = 0.1
x_0 = 1.6

print(f'Gradient descent x_min: {gradient_descent(dfdx_example_1, x_0, learning_rate, num_iterations):.2f}')
```

```python
num_iterations = 75
learning_rate = 0.01
x_0 = 1.6

gd_example_1 = gradient_descent_one_variable(
    [0.001, 2.5],
    [-0.3, 13],
    f_example_1,
    dfdx_example_1,
    gradient_descent,
    num_iterations,
    learning_rate,
    x_0,
    0.0,
    [0.35, 9.5]
)
```
